```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
chipo = pd.read_csv('chipotle.csv',sep='\t')

print(type(chipo))
chipo
```

```python
cnt=len(chipo['order_id'].unique())+1

df_tmp=chipo[['order_id','item_name']]
df_tmp_arr=[[]for i in range(cnt)]
num=0
for i in df_tmp['item_name'] :
    df_tmp_arr[df_tmp['order_id'][num]].append(i)
    num+=1

# df_tmp_arr 리스트화 한것 
```

```python
df_tmp_arr.pop(0)
num=0
for i in df_tmp_arr :
    df_tmp_arr[num] = list(set(df_tmp_arr[num])) #중복제거 연관에서는 개수가 필요 x
    num+=1
df_tmp_arr[:5]
```

```python
df_tmp_arr[:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(df_tmp_arr).transform(df_tmp_arr)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
## 연관도 분석
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
import pandas as pd
df=pd.read_csv('CUST_ORDER_OUTPUT2.utf8.csv')
df.info()
```

```python
def colSpecialChar(df): # 유니크값 확인
    for x in df.columns:
        if df[x].dtype=='object':
            print('--------[%s] data set unique --------' % x)
            print(df[x].unique())
```

```python
colSpecialChar(df)
# sex * 있어 제거
# 나이도 *과 숫자가 문자열과 정수형태로 나누워져 있다
# date도 *있다
```
