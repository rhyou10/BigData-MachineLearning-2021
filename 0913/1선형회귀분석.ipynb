## 선형식

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [20, 35, 20 , 20 ,40]
w=[1,2,3,4,5]; b=[1,3,2,2,3]

for w1,b1 in zip(w,b): #인공지능의 경우

    y_hat = np.array(x)*w1+b1
    print(y_hat)
    plt.plot(x,y,'ro') #점으로
    plt.plot(x,y_hat)  # 계산한거
    #plt.show()
```

## 인공신경망으로 보는 선형식

```python
x = [1, 2, 3, 4, 5]
y = [20, 35, 20 , 20 ,40]
w=0.3; b=0.2

for epoch in range(10): #인공신경망에서 자료 몇번 돌릴지 반복획수 epoch 라고함
    w *=2; b*=2 #인공신경망에서는 w와 b를 오차역전파 수식을 이용하여 업데이트함
    y_hat = np.array(x)*w+b # 예측분류에 따라 y_hat을 구하는 공식이 다름
    print(y_hat)
    plt.plot(x,y,'ro') #점으로
    plt.plot(x,y_hat)  # 계산한거
    #plt.show() #너무많이 나와 생략
```

## 오차
- err = y - y_hat
- ae와 se 가 있다
- 절대값을 씌우는경우 제곱을 하는 경우 있다
- Mse = avg(se) , RMSE = root(avg(se))

```python
x = [1, 2, 3, 4, 5]
y = [20, 35, 20 , 20 ,40]
w=0.3; b=0.2; le = 0.01; err =1;

for epoch in range(10): #인공신경망에서 자료 몇번 돌릴지 반복획수 epoch 라고함
    w =w*err*le; b=b*err*le #인공신경망에서는 w와 b를 오차역전파 수식을 이용하여 업데이트함
    y_hat = np.array(x)*w+b # 예측분류에 따라 y_hat을 구하는 공식이 다름

    err = np.mean(np.sum(y-y_hat)) # 에측분류에 따라 오차값을 구하는 공식이 다르다
    #여기서 중요한점 인공신경망에서는 오차값을 통해 w,b 를 수정 반영한다

    print(y_hat)
    print(err)
    
    plt.plot(x,y,'ro') #점으로
    plt.plot(x,y_hat)  # 계산한거
    #plt.show() #너무많이 나와 생략

```

## batch
- 오차값이 하나가 큰경우 대비하여 이상치 잘 만들것

```python
x = [1, 2, 3, 4, 5]
y = [20, 35, 20 , 20 ,40]
w=0.3; b=0.2; le = 0.01; err =1;
batch=2 #batch가 1일때 좋다
for epoch in range(10): #인공신경망에서 자료 몇번 돌릴지 반복획수 epoch 라고함
    

    for i in range(len(x)/batch):
        w =w*err*le; b=b*err*le #인공신경망에서는 w와 b를 오차역전파 수식을 이용하여 업데이트함
        y_hat = np.array(x[0:batch])*w+b # 예측분류에 따라 y_hat을 구하는 공식이 다름
        # 오차를 줄이기 y x를 batch 잘라서 예측
        # 10 번을 하고 참고하느냐 한번 하고 참고하느냐(오래걸린다)


        err = np.mean(np.sum(y-y_hat)) # 에측분류에 따라 오차값을 구하는 공식이 다르다
        #여기서 중요한점 인공신경망에서는 오차값을 통해 w,b 를 수정 반영한다

        print(y_hat)
        print(err)
```
